### Step 1: Read the dataset

```python
import pandas as pd 
df = pd.read_csv('iris.csv')
df.head()
```

### Step 2: Perform basic data quality checks

```python
df.info()
```

```python
df.isna().sum()
```

```python
df.duplicated().sum()
```

### Step 3: Seperate X and Y(species)

```python
X = df.drop(columns=['species'])
Y = df[['species']]
```

```python
X.head()
```

```python
Y.head()
```

```python
Y.value_counts()
```

```python
Y.value_counts().plot(kind='bar')
```

### Step 4: Apply preprocessing on X

```python
from sklearn.pipeline import Pipeline
from sklearn.impute import SimpleImputer
from sklearn.preprocessing import StandardScaler
```

```python
num_pipe = Pipeline(steps=[('impute', SimpleImputer(strategy='mean')),
                           ('scaler', StandardScaler())]).set_output(transform='pandas')
```

```python
X_pre = num_pipe.fit_transform(X)
X_pre.head()
```

### Step 5: Apply Train test split

```python
from sklearn.model_selection import train_test_split
xtrain, xtest, ytrain, ytest = train_test_split(X_pre, Y, test_size=0.33,
                                                random_state=42)
```

```python
xtrain.shape
```

```python
xtest.shape
```

```python
xtrain.head()
```

```python
ytrain.head()
```

```python
xtest.head()
```

```python
ytest.head()
```

### Step 6 : Model Building

```python
from sklearn.tree import DecisionTreeClassifier
dtc = DecisionTreeClassifier(max_depth=1, 
                             min_samples_split=10, 
                             min_samples_leaf=1,
                             criterion='gini')
dtc.fit(xtrain, ytrain)
```

```python
dtc.score(xtrain, ytrain)
```

```python
dtc.score(xtest, ytest)
```

### Step 7: Hyperparameter tuning

```python
params = {'max_depth':[1, 2, 3, 4, 5],
          'min_samples_split':[1, 2, 3, 4, 5, 6, 7],
          'min_samples_leaf':[1, 2, 3, 4, 5, 6, 7],
          'criterion':['gini','entropy']}
```

```python
from sklearn.model_selection import GridSearchCV
dtc2 = DecisionTreeClassifier()
gscv = GridSearchCV(dtc2, param_grid=params, cv=5, scoring='f1_macro')
gscv.fit(xtrain, ytrain)
```

```python
gscv.best_params_
```

```python
gscv.best_score_
```

```python
best_dtc = gscv.best_estimator_
best_dtc
```

### Step 8: Evaluate the model

```python
best_dtc.score(xtrain, ytrain)
```

```python
best_dtc.score(xtest, ytest)
```

```python
ypred_train = best_dtc.predict(xtrain)
ypred_test = best_dtc.predict(xtest)
```

```python
ypred_train[0:5]
```

```python
ytrain.head()
```

```python
ypred_test[0:5]
```

```python
ytest.head()
```

### Plot confusion matrix for test data

```python
best_dtc.classes_
```

```python
from sklearn.metrics import confusion_matrix, ConfusionMatrixDisplay
cf = confusion_matrix(ytest, ypred_test)
cfd = ConfusionMatrixDisplay(cf, display_labels=best_dtc.classes_)
cfd.plot()
```

### Check classification report for test

```python
from sklearn.metrics import classification_report
print(classification_report(ytest, ypred_test))
```

```python
from sklearn.model_selection import cross_val_score 
f1_scores = cross_val_score(best_dtc, xtrain, ytrain, cv=5, scoring='f1_macro')
f1_scores
```

```python
f1_scores.mean()
```

```python
acc_scores = cross_val_score(best_dtc, xtrain, ytrain, cv=5, scoring='accuracy')
acc_scores
```

```python
acc_scores.mean()
```

### Plotting the decision tree

```python
import matplotlib.pyplot as plt
from sklearn.tree import plot_tree
plt.figure(figsize=(10, 10))
plot_tree(best_dtc, 
          feature_names=xtrain.columns, 
          class_names=best_dtc.classes_,
          filled=True)
plt.show()
```

### Step 9: Out of sample predictions

```python
xnew = pd.read_csv('iris_sample.csv')
xnew
```

```python
xnew_pre = num_pipe.transform(xnew)
xnew_pre
```

```python
preds = best_dtc.predict(xnew_pre)
preds
```

```python
probs = best_dtc.predict_proba(xnew_pre)
probs[0:5]
```

```python
import numpy as np 
max_probs = np.max(probs, axis=1)
max_probs
```

### Save above to xnew

```python
xnew['species_pred'] = preds 
xnew['prob'] = max_probs
```
