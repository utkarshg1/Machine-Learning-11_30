### Problem statement : Segment the customers on basis of Age, Income and Spending

### Step 3: Sepeate X (Unsupervised machine learning does not have target feature)

```python
X = df[['Age', 'Income', 'Spending']]
X.head()
```

### Step 4: Create a preprocessing pipeline for X

```python
from sklearn.pipeline import Pipeline
from sklearn.impute import SimpleImputer
from sklearn.preprocessing import StandardScaler
```

```python
num_pipe = Pipeline(steps=[('impute', SimpleImputer(strategy='median')),
                           ('scaler', StandardScaler())]).set_output(transform='pandas')
```

```python
X_pre = num_pipe.fit_transform(X)
```

```python
X_pre.head()
```

### Step 5: Find optimal number of clusters for given data

```python
from sklearn.metrics import silhouette_score
```

```python
n = list(range(2, 15))
n
```

```python
from sklearn.cluster import AgglomerativeClustering
sil = []
for i in n:
    model = AgglomerativeClustering(n_clusters=i)
    clusters = model.fit_predict(X_pre)
    s = silhouette_score(X_pre, clusters)
    sil.append(s)
```

```python
sil
```

### Plotting silhoutee scores

```python
import matplotlib.pyplot as plt
plt.figure(figsize=(10,5))
plt.plot(n, sil)
plt.xlabel('Number of clusters')
plt.ylabel('Silhoutee scores')
plt.title('Silhoutee scores for different clusters')
plt.grid()
plt.show()
```

### From above plot i have highest silhoutee score at n=6

```python
agc_model = AgglomerativeClustering(n_clusters=6)
clusters = agc_model.fit_predict(X_pre)
clusters
```

## Save above in X 

```python
X['Cluster'] = clusters
```

```python
X
```
