### Step 1: Read the train dataset

```python
import pandas as pd 
df = pd.read_csv('train_titanic.csv')
df.head()
```

### Step 2: Perform basic data quality checks

```python
df.info()
```

```python
m = df.isna().sum()
m
```

```python
df.duplicated().sum()
```

### Step 3: Seperate X and Y(Survived)

```python
X = df.drop(columns=['PassengerId', 'Name', 'Ticket', 'Cabin', 'Survived'])
Y = df[['Survived']]
```

```python
X.head()
```

```python
Y.head()
```

```python
Y.value_counts()
```

```python
Y.value_counts().plot(kind='bar')
```

### Step 4: Preprocess data for X

```python
# cat con seperation
cat = list(X.columns[X.dtypes=='object'])
con = list(X.columns[X.dtypes!='object'])
```

```python
cat
```

```python
con
```

```python
from sklearn.pipeline import Pipeline
from sklearn.impute import SimpleImputer
from sklearn.preprocessing import StandardScaler, OneHotEncoder
from sklearn.compose import ColumnTransformer
```

```python
num_pipe = Pipeline(steps=[('impute', SimpleImputer(strategy='median')),
                           ('scaler', StandardScaler())])
```

```python
cat_pipe = Pipeline(steps=[('impute', SimpleImputer(strategy='most_frequent')),
                           ('ohe', OneHotEncoder(handle_unknown='ignore', sparse_output=False))])
```

```python
pre = ColumnTransformer([('num', num_pipe, con),
                         ('cat', cat_pipe, cat)]).set_output(transform='pandas')
```

```python
X_pre = pre.fit_transform(X)
X_pre.head()
```

### Step 5: Apply Train Test Split

```python
from sklearn.model_selection import train_test_split
xtrain, xtest, ytrain, ytest = train_test_split(X_pre, Y, test_size=0.2, random_state=42)
```

```python
xtrain.head()
```

```python
ytrain.head()
```

```python
xtest.head()
```

```python
ytest.head()
```

```python
xtrain.shape
```

```python
xtest.shape
```

### Step 6: Model Building (Random Forest)

```python
from sklearn.ensemble import RandomForestClassifier
model1 = RandomForestClassifier(n_estimators=1, 
                                max_depth=1, 
                                min_samples_split=5,
                                min_samples_leaf=5,
                                criterion='entropy')
model1.fit(xtrain, ytrain)
```

```python
# Accuracy in train
model1.score(xtrain, ytrain)
```

```python
# Accuracy in Test
model1.score(xtest, ytest)
```

### Hyperparameter tuning

```python
params1 = {'n_estimators':[10, 50, 100, 200, 300],
           'max_depth':[2, 3, 4, 5],
           'min_samples_split':[5, 6, 7, 8],
           'criterion':['gini', 'entropy']}
```

```python
# RandomizedSearch is faster than gridsearch
from sklearn.model_selection import RandomizedSearchCV
rfc = RandomForestClassifier()
rscv1 = RandomizedSearchCV(rfc, param_distributions=params1, cv=5, scoring='f1_macro')
rscv1.fit(xtrain, ytrain)
```

```python
rscv1.best_params_
```

```python
rscv1.best_score_
```

```python
best_rfc = rscv1.best_estimator_
best_rfc
```

```python
best_rfc.score(xtrain, ytrain)
```

```python
best_rfc.score(xtest, ytest)
```

### Predict results for random forest

```python
ypred_train1 = best_rfc.predict(xtrain)
ypred_test1 = best_rfc.predict(xtest)
```

```python
ypred_train1[0:5]
```

```python
ytrain.head()
```

```python
ypred_test1[0:5]
```

```python
ytest.head()
```

### Plot confusion Matrix for random forest

```python
from sklearn.metrics import confusion_matrix, ConfusionMatrixDisplay
cf1 = confusion_matrix(ytest, ypred_test1)
cfd1 = ConfusionMatrixDisplay(cf1, display_labels=best_rfc.classes_)
cfd1.plot()
```

### Print classification report for RandomForest

```python
from sklearn.metrics import classification_report 
print(classification_report(ytest, ypred_test1))
```

### Above model is a good model

# Gradient Boosting Model

```python
from sklearn.ensemble import GradientBoostingClassifier
model2 = GradientBoostingClassifier(learning_rate=0.1,
                                    n_estimators=1,
                                    max_depth=1,
                                    min_samples_split=5, 
                                    min_samples_leaf=5)
model2.fit(xtrain, ytrain)
```

```python
model2.score(xtrain, ytrain)
```

```python
model2.score(xtest, ytest)
```

### Hyperparameter tuning with Randomised search cv

```python
params2 = {'learning_rate':[0.01, 0.1, 0.15, 0.2],
           'n_estimators':[10, 50, 100, 200, 300],
           'max_depth':[2, 3, 4, 5],
           'min_samples_split':[5, 6, 7, 8, 9]}
```

```python
gbc = GradientBoostingClassifier()
rscv2 = RandomizedSearchCV(gbc, param_distributions=params2, cv=5, scoring='f1_macro')
rscv2.fit(xtrain, ytrain)
```

```python
rscv2.best_params_
```

```python
rscv2.best_score_
```

```python
best_gbc = rscv2.best_estimator_
best_gbc
```

```python
best_gbc.score(xtest, ytest)
```
