# Feature Selection - Forward Feature selection
## Feature Selection means selecting only specific columns from given data for Machine Learning

```python
from warnings import filterwarnings
filterwarnings('ignore')
```

```python
import pandas as pd
df = pd.read_csv('Cars93.csv', na_values=['','NA'], keep_default_na=False)
df.head()
```

### Perform basic data quality check

```python
df.info()
```

```python
m = df.isna().sum()
m[m>0]
```

```python
df.duplicated().sum()
```

### Seperate X and Y (Weight)

```python
X = df.drop(columns=['id', 'Weight'])
Y = df[['Weight']]
```

```python
X.head()
```

```python
Y.head()
```

### Sepearate cat and con features for X

```python
cat = list(X.columns[X.dtypes=='object'])
con = list(X.columns[X.dtypes!='object'])
```

```python
cat
```

```python
con
```

### Create a feature selection pipeline
(Categorical features - Ordinal Encoding)

```python
from sklearn.pipeline import Pipeline
from sklearn.impute import SimpleImputer
from sklearn.preprocessing import StandardScaler, OrdinalEncoder
from sklearn.compose import ColumnTransformer
```

```python
num_pipe1 = Pipeline(steps=[('imputer', SimpleImputer(strategy='mean')),
                            ('scaler', StandardScaler())])
```

```python
cat_pipe1 = Pipeline(steps=[('impute', SimpleImputer(strategy='most_frequent')),
                            ('ordinal', OrdinalEncoder())])
```

```python
pre1 = ColumnTransformer([('num', num_pipe1, con),
                          ('cat', cat_pipe1, cat)]).set_output(transform='pandas')
```

```python
pre1
```

```python
X_pre = pre1.fit_transform(X)
X_pre.head()
```

### Feature selection

```python
from sklearn.linear_model import LinearRegression
from sklearn.feature_selection import SequentialFeatureSelector
```

```python
lr = LinearRegression()
sel = SequentialFeatureSelector(lr, direction='forward')
sel.fit_transform(X_pre, Y)
sel_cols = sel.get_feature_names_out()
sel_cols
```

```python
len(sel_cols)
```

```python
sel_cols[0]
```

```python
sel_cols[0].split('__')
```

```python
sel_cols[0].split('__')[1]
```

```python
imp_cols = []
for i in sel_cols:
    s = i.split('__')[1]
    imp_cols.append(s)
```

```python
imp_cols
```

### Get above dataframe

```python
X_sel = X[imp_cols]
X_sel
```

### Seperate cat and con features from X_sel

```python
cat_sel = list(X_sel.columns[X_sel.dtypes=='object'])
con_sel = list(X_sel.columns[X_sel.dtypes!='object'])
```

```python
cat_sel
```

```python
con_sel
```

### Create a final pipeline for model building
cateogrical - OneHotEncoded

```python
from sklearn.preprocessing import OneHotEncoder
```

```python
num_pipe2 = Pipeline(steps=[('impute', SimpleImputer(strategy='mean')),
                            ('scaler', StandardScaler())])
```

```python
cat_pipe2 = Pipeline(steps=[('impute', SimpleImputer(strategy='most_frequent')),
                            ('ohe', OneHotEncoder(handle_unknown='ignore', sparse_output=False))])
```

```python
pre2 = ColumnTransformer([('num', num_pipe2, con_sel),
                          ('cat', cat_pipe2, cat_sel)]).set_output(transform='pandas')
```

```python
X_sel_pre = pre2.fit_transform(X_sel)
X_sel_pre
```

### Apply the train test split

```python
from sklearn.model_selection import train_test_split
xtrain, xtest, ytrain, ytest = train_test_split(X_sel_pre, Y, test_size=0.2, random_state=21)
```

```python
xtrain.shape
```

```python
xtest.shape
```

### Build the model

```python
from sklearn.linear_model import LinearRegression
model = LinearRegression()
model.fit(xtrain, ytrain)
```

```python
model.score(xtrain, ytrain)
```

```python
model.score(xtest, ytest)
```
